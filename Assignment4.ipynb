```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

cols = ['Item','Category', 'Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)']
df = pd.read_csv('menu.csv', usecols=cols)
df.describe()
```

```python
df['Category']
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x='Calories', y='Category', data=df)
plt.title('Box Plot between Calories and Category')
plt.xlabel('Calories')
plt.ylabel('Category')
plt.show()
```

```python
df.set_index('Item')
```

```python
def high_q_item(item, metric, df, Item, Quantity):
    ref = df[metric].idxmax()
    Item.append(df.iloc[ref][item])  
    Quantity.append(df.iloc[ref][metric])

    return Item, Quantity     
```

```python
Item = []
Quantity = []
metrics = df.columns
metrics = metrics.drop(['Category', 'Item'])
for c in metrics:
    Item, Quantity = high_q_item('Item', c, df, Item, Quantity)

for item, quantity, c in zip(Item, Quantity,metrics):
    print(f"{c}:  {item}  -  {quantity}")
```

# **Task 6**

```python


# Loop through each attribute
for attribute in metrics:
    # Create a strip plot for the current attribute
    plt.figure(figsize=(10, 6))
    sns.stripplot(x='Category', y=attribute, data=df, jitter=True, palette='Set1')
    
    # Set plot title and labels
    plt.title(f'Strip Plot of {attribute} by Category')
    plt.xlabel('Category')
    plt.ylabel(attribute)
    
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45)
    
    # Show plot
    plt.tight_layout()
    plt.show()
```

```python

categories = df['Category'].unique()
bar_width = 0.5

for category in categories:
    # Filter the dataframe for the current category
    category_df = df[df['Category'] == category]
    items = category_df['Item']
    calories = category_df['Calories']

    plt.figure(figsize=(20, 20))
    
    # Create a bar plot for the current category
    plt.barh(items, calories,  label=category)
    
    # Add labels and title
    plt.xlabel('Item')
    plt.ylabel('Calories')
    plt.title(f'Calories of Items in {category}')
    
    # Show plot
    plt.show()
```
